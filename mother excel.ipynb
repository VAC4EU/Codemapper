````python
import polars as pl
import json
````

````python
projects0 = pl.read_csv('VAC4EU PASS CovidVaccineMonitoringVariables_2025_w23(CDM_EVENTS).csv')

name_colunm = 'Event_abbreviation / Variable Name'
descendants_column = 'Descendants_Rule'
project_columns = "GCOVS	RSV-adult	RSV-MATERNAL	Paxlovid_1037	Paxlovid_1047	ROC18	SAFETY-VAC (OBJ3)	Pfizer pregnancy algorithm	Pfizer	Pfizer_Myo	1052	AZ	Moderna	Moderna Pregnancy	Moderna_Myo	J&J	ROC20	ADEPT	CONSIGN	ALL	EMA_ROC11_CONSIGN	EMA_ROC12_effectiveness	Conception	ConcePTIONAlgorithmPregnancies".split('\t')

projects = (
    projects0
    .rename({})[[name_colunm, descendants_column] + project_columns]
    .filter(pl.col(name_colunm).is_not_null())
    .with_columns(pl.col(name_colunm).str.strip_chars())
)
projects
````

````python
uri = "postgresql://localhost:5432/codemapper"
query = "SELECT id, name, meta FROM case_definitions WHERE project_id = 47 AND state is NULL"
mappings = pl.read_database_uri(query=query, uri=uri)
mappings = mappings.with_columns(
    pl.concat_str(
        pl.col('meta').str.json_path_match("$.system"),
        pl.col("name"),
        pl.col("meta").str.json_path_match("$.type"),
        separator='_'
    ).alias(name_colunm)).drop(["name", "meta"])
mappings
````

````python
projects_mappings = mappings.join(projects, on=name_colunm, how="right")
projects_mappings.to_pandas()
````

````python
missing = projects_mappings.filter(pl.col("id").is_null())[name_colunm]
print(len(missing))
missing.str.join(', ')[0]
````

````python
final = (
    projects_mappings
        .unpivot(index=['id', name_colunm, descendants_column], on=project_columns, variable_name='Project')
        .filter(pl.col("value").is_not_null())
        .group_by(['id', name_colunm])
        .agg([
            pl.col("Project").alias("Projects"),
            pl.col(descendants_column).any()
        ])
        .with_columns(pl.col("Projects").map_elements(lambda e: json.dumps(list(e))))
        .with_columns(pl.col("Projects").map_elements(lambda e: json.dumps(list(e))))
        .filter(pl.col("id").is_not_null())
)
final.to_pandas()
````

````python
final.write_csv('mother excel.csv', include_header=False)
````

from mother

    select case_definition_id, max(version) as version
    from case_definition_revisions
    group by case_definition_id
) as latest
on mother.case_definition_id = latest.case_definition_id
where r.case_definition_id = latest.case_definition_id
and r.version = latest.version;
